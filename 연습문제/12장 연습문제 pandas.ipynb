```python
a = df.iloc[:, 1:]

a.sum(axis=1)
```

```python
a1 = a.sum(axis=1)
```

```python
df['의사 수'] = a.sum(axis=1)
df
```

```python
df.plot.scatter(x = '지역', y = '의사 수')

font1 = {'size':18, 'color':'green'}
plt.title('서울 및 광역시 종합병원 전문의 수', fontdict=font1)
plt.xlabel('시')
plt.ylabel('의사 수')
plt.show()
```

```python
5.

xdata = df['지역']
ydata1 = df['일반의 수']
ydata2 = df['인턴 수']
ydata3 = df['레지던트 수']
ydata4 = df['전문의 수']

plt.plot(xdata, ydata1, label='국어', color='red', linestyle='-', marker='o')
plt.plot(xdata, ydata2, label='영어', color='#00ffff', linestyle='--', marker='x')
plt.plot(xdata, ydata3, label='수학', color='magenta', linestyle='-.', marker='s')
plt.plot(xdata, ydata4, label='사회', color='#444444', linestyle=':', marker='d')

plt.title('서울 및 광역시 종합병원 의사 수')
plt.legend(loc='best')
plt.show()
```
