```python
def combinations(s):
    if len(s) < 2 :
        return s
    res = []
    for i, c in enumerate(s) :
        res.append(c)
        #print(i, c, res)       # 작업이 어떻게 돌아가는지 보여줌
        for j in combinations(s[:i] + s[i+1:]) :
            res.append(c+j)
    return res
```

```python
if __name__ == "__main__" :
    result = combinations(input(''))
    print(result)
```
